# Carga de librerías
El código cargará las librerías de numpy, matplotlib y sympy para la generación de la respuesta de la consigna 2

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
x = Symbol('x')
f1 = 5*exp(-10*x**3)
f2 = -100*x**2 +91*x-16.25
f = f1-f2
f_dot = f.diff()
f_np = lambdify(x,f,'numpy')
f1_np = lambdify(x,f1,'numpy')
f2_np = lambdify(x,f2,'numpy')
f_dot_np = lambdify(x,f_dot,'numpy')
```

```python
def NewtonRaphson(p0, tol, f, der_f, No):
    i=1
    salida="No lo logre"
    while i<No:
        p = p0 - ((f(p0))/(der_f(p0)))
        if abs(p-p0) < tol:
            #print("Lo logre. La raíz esta: ", p)
            salida = (p,i)
            break
        i = i+1
        p0 = p
    return salida
```

```python
L2, k = NewtonRaphson(0.6, 1e-5, f_np, f_dot_np, 10000)
```

```python
t = np.linspace(0, 1, 1000)
figura01 = plt.figure()
plt.plot(t, f_np(t))
plt.plot(t, f1_np(t))
plt.plot(t, f2_np(t))
plt.plot(L2, f1_np(L2), 'o', zorder=10)
plt.ylim(-1,1)
plt.grid()
```
