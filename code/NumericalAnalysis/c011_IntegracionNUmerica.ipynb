```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2
```

```python
x = np.linspace(1.5,4.5,1000)
y = f(x)
pto_x = np.array([2,4])
pto_y = f(pto_x)
figura01 = plt.figure()
plt.plot(x,y)
plt.plot(pto_x, pto_y, 'o')
plt.axis([1.5, 4.5, 0, 20])
plt.grid()
```

```python
x1 = np.array([2, 2.25, 2.5,2.75,3,3.25,3.5,3.75,4])
y1 = f(x1)
salida = 0
for k in x1:
    salida = salida + 0.25*f(k)
salida
```

```python
h=0.2
x2 = np.arange(2,4+h,h)
y2 = f(x2)
salida = 0
for k in x2:
    salida = salida + h*f(k)
salida
```
